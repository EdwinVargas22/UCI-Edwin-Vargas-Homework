```python
# Import Dependencies
import pandas as pd
```

```python
# Reference to CSV file
csv_path = "Resources/HeroesOfPymoli_data.csv"

# Read Heroes of Pymoli data and store into Pandas DataFrame
Heroes_df = pd.read_csv(csv_path)
Heroes_df.head()
```

```python
# Removing duplicate players
players_df = Heroes_df.drop_duplicates(subset=["SN"])

# Counting up the total players
total_players_count = players_df.shape[0]

mini_summary_df = pd.DataFrame({"Total Players": [total_players_count]})

mini_summary_df
```

```python
# Finding the number of unique items
number_unique_items = len(Heroes_df["Item ID"].unique())

# Calculating the average price
average_price = Heroes_df["Price"].mean()


# Finding the number of purchases
number_purchases = len(Heroes_df["Purchase ID"])

# Calculating the total revenue
total_revenue = Heroes_df["Price"].sum()
```

```python
# Created a summary DataFrame for Purchasing Analysis(Total)
summary_df = pd.DataFrame({"Number of Unique Items": [number_unique_items],
                              "Average Price": [average_price],
                              "Number of Purchases": [number_purchases],
                              "Total Revenue": [total_revenue]})

# Used Map to format the Average Price and Total Revenue columns
summary_df["Average Price"] = summary_df["Average Price"].map("${:.2f}".format)
summary_df["Total Revenue"] = summary_df["Total Revenue"].map("${:,}".format)

summary_df
```

```python
# Total Count of Gender Demographics
total_gender_count = players_df["Gender"].value_counts()

total_gender_count
```

```python
# Total Count of Percentages per Gender Demographics
total_percentage_players = players_df["Gender"].value_counts(normalize=True)

# Times the percentage by 100
percentage_players = total_percentage_players * 100

percentage_players
```

```python
# Created a summary DataFrame for Percentages and Counts per Gender Demographics
gender_demographics_summary_df = pd.DataFrame({"Total Count": total_gender_count,
                                      "Percentage of Players": percentage_players})

# Used Map to format Percentage of Players column
gender_demographics_summary_df["Percentage of Players"] = gender_demographics_summary_df["Percentage of Players"].map("{:.2f}%".format)

gender_demographics_summary_df
```

```python
# Group by gender from the Heroes DataFrame
grouped_gender_df = Heroes_df.groupby(["Gender"])

# Used calculations 
purchase_count = grouped_gender_df["Purchase ID"].count()

average_purchase = grouped_gender_df["Price"].mean()

total_purchase = grouped_gender_df["Price"].sum()

avg_total_purchase_person = total_purchase/total_gender_count

# Created a summary DataFrame for Purchasing Analysis (Gender)
purchasing_analysis_summary_df = pd.DataFrame({"Purchase Count": purchase_count,
                                              "Average Purchase Price": average_purchase,
                                              "Total Purchase Value": total_purchase,
                                              "Avg Total Purchase per Person": avg_total_purchase_person})

# Used Map to format three columns
purchasing_analysis_summary_df["Average Purchase Price"] = purchasing_analysis_summary_df["Average Purchase Price"].map("${:.2f}".format)
purchasing_analysis_summary_df["Total Purchase Value"] = purchasing_analysis_summary_df["Total Purchase Value"].map("${:,.2f}".format)
purchasing_analysis_summary_df["Avg Total Purchase per Person"] = purchasing_analysis_summary_df["Avg Total Purchase per Person"].map("${:.2f}".format)



purchasing_analysis_summary_df
```

```python
# Group by SN from the Heroes DataFrame
group_sn = Heroes_df.groupby(["SN"])

# Used Calculations
sn_purchase_count = group_sn["Purchase ID"].count()

sn_average_price = group_sn["Price"].mean()

sn_total_price = group_sn["Price"].sum()

# Created a summary DataFrame for Top Spenders
top_spenders_summary_df = pd.DataFrame({"Purchase Count": sn_purchase_count,
                                              "Average Purchase Price": sn_average_price,
                                              "Total Purchase Value": sn_total_price})

# Sort the total purchase value to descending order and set the summary DataFrame to a new DataFrame
top_spenders_df = top_spenders_summary_df.sort_values("Total Purchase Value", ascending=False)

# Use style.format on the average purchase price and total purchase value so it could be set as a dollar value and float
top_spenders_df.head().style.format({"Average Purchase Price": "${:,.2f}", "Total Purchase Value": "${:,.2f}"})
```

```python
# dtypes of top spenders DataFrame
top_spenders_df.dtypes
```

```python
# Retrieved the Item ID, Item Name, and Price columns to create a new DataFrame 
retrieve_df = Heroes_df[["Item ID", "Item Name", "Price"]]

# Group by Item ID and Item Name from the retrieve DataFrame
group_names = retrieve_df.groupby(["Item ID", "Item Name"])

# Used Calculations
name_purchase_count = group_names["Price"].count()

name_average_price = group_names["Price"].mean()

name_total_price = group_names["Price"].sum()

# Created a summary DataFrame for Most Popular Items
most_popular_items_summary_df = pd.DataFrame({"Purchase Count": name_purchase_count,
                                              "Item Price": name_average_price,
                                              "Total Purchase Value": name_total_price})

# Sort the purchase count to descending order and set the summary DataFrame to a new DataFrame
popular_items_df = most_popular_items_summary_df.sort_values("Purchase Count", ascending=False)


# Use style.format on the item price and total purchase value so it could be set as a dollar value and float
popular_items_df.head().style.format({"Item Price": "${:,.2f}", "Total Purchase Value": "${:,.2f}"})
```

```python
# Finding the Most Profitable Items
# Sort the total purchase value to descending order and set the summary DataFrame to a new DataFrame
profitable_items_df = most_popular_items_summary_df.sort_values("Total Purchase Value", ascending=False)

# Use style.format on the item price and total purchase value so it could be set as a dollar value and float
profitable_items_df.head().style.format({"Item Price": "${:,.2f}", "Total Purchase Value": "${:,.2f}"})
```
